# 4. What are the model parameters of LDA and QDA classifiers? What is the role of each parameter?Between QDA and LDA, which one has potentially more parameters? why?

### Model Parameters:

#### LDA (Linear Discriminant Analysis):
1. **Mean Vectors (μ):** Represent the mean of each class in the feature space.
2. **Shared Covariance Matrix (Σ):** Assumes that all classes share the same covariance matrix, i.e., assumes equal covariance among classes.

#### QDA (Quadratic Discriminant Analysis):
1. **Mean Vectors (μ):** Same as LDA, representing the mean of each class in the feature space.
2. **Class-Specific Covariance Matrices (Σ):** Allows each class to have its own covariance matrix, making the decision boundary more flexible and potentially better fitting to the data.

### Role of Each Parameter:

- **Mean Vectors (μ):** Determine the central location of each class within the feature space.
- **Covariance Matrices (Σ):** Capture the spread and orientation of data within each class. LDA assumes a shared covariance matrix among classes, while QDA allows separate covariance matrices for each class.

### Comparison of Parameter Count:

- **LDA** has fewer parameters compared to **QDA** because it assumes a shared covariance matrix across all classes.
- **QDA** potentially has more parameters as it estimates a separate covariance matrix for each class, providing more flexibility in capturing distinct covariance structures.

### Potential Explanation:

QDA potentially has more parameters because it estimates a separate covariance matrix for each class. In scenarios with multiple classes or high-dimensional data, QDA's flexibility increases the parameter count compared to LDA, where the assumption of a shared covariance matrix reduces the overall parameter count.

## functions

```python
def mahalanobis_distance(x, mean, covariance):
    diff = x - mean

    distance = np.sqrt(np.dot(np.dot(diff, np.linalg.inv(covariance)), diff.T))

    return distance
```

## LDA Implementation:

```python
class LDA:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.class_means = []
        self.shared_covariance = np.zeros((X.shape[1], X.shape[1]))

        for c in self.classes:
            X_class = X[y == c]
            mean = np.mean(X_class, axis=0)
            self.class_means.append(mean)
            self.shared_covariance += np.cov(X_class, rowvar=False)

        self.shared_covariance /= len(self.classes)

    def predict(self, X):
        predictions = []
        for x in X:
            distances = []
            for _, mean in enumerate(self.class_means):
                distance = mahalanobis_distance(x, mean, self.shared_covariance)
                distances.append(distance)

            predictions.append(self.classes[np.argmin(distances)])

        return np.array(predictions)
    
    def get_parameters_count(self, X, y):
        n_features = X.shape[1]
        n_classes = len(np.unique(y))
        lda_params_count = n_features * n_classes + n_features**2
        return lda_params_count
```

## QDA Implementation:

```python
class QDA:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.class_means = []
        self.class_covariances = []

        for c in self.classes:
            X_class = X[y == c]
            mean = np.mean(X_class, axis=0)
            self.class_means.append(mean)
            self.class_covariances.append(np.cov(X_class, rowvar=False))

    def predict(self, X):
        predictions = []
        for x in X:
            distances = []
            for idx, mean in enumerate(self.class_means):
                distance = mahalanobis_distance(x, mean, self.class_covariances[idx])
                distances.append(distance)

            predictions.append(self.classes[np.argmin(distances)])

        return np.array(predictions)
    
    def get_parameters_count(self, X, y):
        n_features = X.shape[1]
        n_classes = len(np.unique(y))
        qda_params_count = n_features * n_classes + n_classes * n_features**2
        return qda_params_count
```

# 7. Singular Covariance Matrix in Feature Matrices

A singular matrix is a square matrix that does not have an inverse. Put simply, for a matrix to be invertible, it must possess a unique inverse matrix that, when multiplied with the original matrix, results in an identity matrix.

### Causes of a Singular Covariance Matrix:

1. **Redundant or Linearly Dependent Features:**
   - Multicollinearity arises when one or more features are a linear combination of others.
   - Linear dependence causes the covariance matrix to lose rank.
   - Instances include identical or nearly identical features or one feature being a constant multiple of another, leading to singularity.

2. **High Dimensionality with Limited Samples:**
   - In high-dimensional spaces (more features) with few samples, the covariance matrix can become singular.
   - When the number of features approximates or exceeds the sample count, singularity may occur.

3. **Noise or Outliers:**
   - Outliers or noise in the dataset can impact covariance matrix calculation.
   - Presence of outliers, especially in small datasets, can significantly influence estimated covariance, potentially leading to singularity.

# 10. Report accuracy, precision, recall, f1-score and the confusion matrix on both training and test sets for both LDA and QDA models. Which classifier performs better on mushroom dataset?


```python
onfusion_matrix = np.zeros((2, 2), dtype=int)
```

```python
def calculate_accuracy(y_true, y_pred):
    return sum(y_true == y_pred) / len(y_true)


def calculate_confusion_matrix(y_true, y_pred):
    unique_labels = np.unique(np.concatenate((y_true, y_pred)))
    num_labels = len(unique_labels)
    confusion_matrix = np.zeros((num_labels, num_labels), dtype=int)

    for i in range(len(y_true)):
        confusion_matrix[y_true[i], y_pred[i]] += 1

    return confusion_matrix


def calculate_precision_recall_f1(confusion_matrix):
    diagonal = confusion_matrix.diagonal()
    precision = diagonal / confusion_matrix.sum(axis=0)
    recall = diagonal / confusion_matrix.sum(axis=1)
    f1_score = 2 * (precision * recall) / (precision + recall)

    return precision, recall, f1_score
```

```python

lda_predictions_train = lda.predict(X_train.to_numpy())
lda_predictions_test = lda.predict(X_test.to_numpy())

qda_predictions_train = qda.predict(X_train.to_numpy())
qda_predictions_test = qda.predict(X_test.to_numpy())


for model_name, y_true_train, y_pred_train, y_true_test, y_pred_test in [
    (
        "LDA - Training",
        y_train.to_numpy(),
        lda_predictions_train,
        y_test.to_numpy(),
        lda_predictions_test,
    ),
    (
        "LDA - Test",
        y_test.to_numpy(),
        lda_predictions_test,
        y_test.to_numpy(),
        lda_predictions_test,
    ),
    (
        "QDA - Training",
        y_train.to_numpy(),
        qda_predictions_train,
        y_test.to_numpy(),
        qda_predictions_test,
    ),
    (
        "QDA - Test",
        y_test.to_numpy(),
        qda_predictions_test,
        y_test.to_numpy(),
        qda_predictions_test,
    ),
]:
    print(f"Evaluation metrics for {model_name}:")
    if "Training" in model_name:
        y_true, y_pred = y_true_train, y_pred_train
    else:
        y_true, y_pred = y_true_test, y_pred_test

    accuracy = calculate_accuracy(y_true, y_pred)
    print(f"Accuracy: {accuracy:.4f}")

    confusion_mat = calculate_confusion_matrix(y_true, y_pred)
    print("Confusion Matrix:")
    print(confusion_mat)
    precision, recall, f1_score = calculate_precision_recall_f1(confusion_mat)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1-score:", f1_score)
    print("\n")
```

# 11. According to your confusion matrices, how many samples are misclassified in each classifier? Are these misclassified samples as FP (False Positive) or FN (False Negative)? Which one is more important in the task of classifying poisonous mushrooms?


```python
def calculate_misclassified_samples(confusion_matrix):
    return np.sum(confusion_matrix) - np.trace(confusion_matrix)


def calculate_false_positives(confusion_matrix, label_index):
    false_positives = (
        np.sum(confusion_matrix[:, label_index])
        - confusion_matrix[label_index, label_index]
    )
    return false_positives


def calculate_false_negatives(confusion_matrix, label_index):
    false_negatives = (
        np.sum(confusion_matrix[label_index, :])
        - confusion_matrix[label_index, label_index]
    )
    return false_negatives
```

```python
for model_name, y_true_train, y_pred_train, y_true_test, y_pred_test in [
    (
        "LDA - Training",
        y_train.to_numpy(),
        lda_predictions_train,
        y_test.to_numpy(),
        lda_predictions_test,
    ),
    (
        "LDA - Test",
        y_test.to_numpy(),
        lda_predictions_test,
        y_test.to_numpy(),
        lda_predictions_test,
    ),
    (
        "QDA - Training",
        y_train.to_numpy(),
        qda_predictions_train,
        y_test.to_numpy(),
        qda_predictions_test,
    ),
    (
        "QDA - Test",
        y_test.to_numpy(),
        qda_predictions_test,
        y_test.to_numpy(),
        qda_predictions_test,
    ),
]:
    if "Training" in model_name:
        y_true, y_pred = y_true_train, y_pred_train
    else:
        y_true, y_pred = y_true_test, y_pred_test

    confusion_mat = calculate_confusion_matrix(y_true, y_pred)
    misclassified = calculate_misclassified_samples(confusion_mat)

    print(f"Evaluation metrics for {model_name}:")
    print("Confusion Matrix:")
    print(confusion_mat)

    label_index = 1  # Assuming 'poisonous' is the second class (index 1)
    false_positives = calculate_false_positives(confusion_mat, label_index)
    false_negatives = calculate_false_negatives(confusion_mat, label_index)

    print(f"False Positives: {false_positives}")
    print(f"False Negatives: {false_negatives}")

    if false_negatives > false_positives:
        print("False Negatives are more critical in classifying poisonous mushrooms.")
    elif false_positives > false_negatives:
        print("False Positives are more critical in classifying poisonous mushrooms.")
    else:
        print("False Positives and False Negatives have equal importance.")

    print(f"Misclassified samples: {misclassified}")
    print("\n")
```

```python
def calculate_class_priors(y):
    _, counts = np.unique(y, return_counts=True)
    return counts / len(y)


def calculate_class_means(X, y):
    class_means = []
    for c in np.unique(y):
        X_class = X[y == c]
        mean = np.mean(X_class, axis=0)
        class_means.append(mean)
    return np.array(class_means)


def calculate_shared_covariance(X, y):
    class_means = calculate_class_means(X, y)
    S_W = np.zeros((X.shape[1], X.shape[1]))
    for c, mean in zip(np.unique(y), class_means):
        X_class = X[y == c]
        S_W += np.cov(X_class, rowvar=False) * len(X_class)
    return S_W / len(y)
```
